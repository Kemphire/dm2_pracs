```python
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

X, Y = make_classification(
    n_classes=2, n_samples=1000, random_state=34, n_features=10, n_informative=7
)

X_train,X_test,y_train,y_test = train_test_split(X,Y,train_size=0.75)
```

# Bagging Classifier Cosnsisting of trees

### 3 Trees

```python
clf = BaggingClassifier(n_estimators=3,estimator=DecisionTreeClassifier())

clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)

conf_mat_bag_3 = confusion_matrix(y_true=y_test,y_pred=y_pred)
```

### 5 Trees

```python
clf = BaggingClassifier(n_estimators=5,estimator=DecisionTreeClassifier())

clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)

conf_mat_bag_5 = confusion_matrix(y_true=y_test,y_pred=y_pred)
```

### 7 Trees

```python
clf = BaggingClassifier(n_estimators=7,estimator=DecisionTreeClassifier())

clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)

conf_mat_bag_7 = confusion_matrix(y_true=y_test,y_pred=y_pred)
```

### 9 Trees

```python
clf = BaggingClassifier(n_estimators=9,estimator=DecisionTreeClassifier())

clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)

conf_mat_bag_9 = confusion_matrix(y_true=y_test,y_pred=y_pred)
```

## Comparing Bagging classifier

```python
import matplotlib.pyplot as plt
import seaborn as sns

confusion_matrices = [conf_mat_bag_3,conf_mat_bag_5,conf_mat_bag_7,conf_mat_bag_9]

fig,axes = plt.subplots(1,4,figsize=(28,7))

model_names = ["conf_mat_bag_3","conf_mat_bag_5","conf_mat_bag_7","conf_mat_bag_9"]

for ax,cm,name in zip(axes,confusion_matrices,model_names):
    sns.heatmap(cm,annot=True,ax=ax)
    ax.set_title(name)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Actual")

plt.title("Comparision of Bagging Classifier")
plt.show()
```
